```
(
    df
 .groupby('param_marka-pojazdu')['price_value']
 .agg(np.mean)
 .sort_values(ascending=False) #sortowanie po wartościach malejące
 .head(50) # head zostawia 50 pierwszych pozycji

) .plot(kind='bar', figsize=(15, 5)) # figsize dodane w plot pozwala rozciągnać wykres - w tym przypadku w stosunku 3:1
#";" na końcu usuwa pierwszy rząd tekstu widocznego poniżej - zobcza przykład wyżej
```

```
(
    df
 .groupby('param_marka-pojazdu')['price_value']
 .agg(np.median) #mediana zmienia ułożenie wykresu - jest odporna na wartości odstające
 .sort_values(ascending=False)
 .head(50)

) .plot(kind='bar', figsize=(15, 5))
```

```
(
    df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.median, np.mean, np.size])
 .sort_values(by='mean', ascending=False)
 .head(50)

) .plot(kind='bar', figsize=(15, 5))
```

```
(
    df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.median, np.mean, np.size])
 .sort_values(by='mean', ascending=False)
 .head(50)

) .plot(kind='bar', figsize=(20, 8), subplots=True) # subplots=True pozwala rozdzielić wykresy
```

```
(
    df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.median, np.mean, np.size])
 .sort_values(by='size', ascending=False)
 .head(50)

) .plot(kind='bar', figsize=(20, 8), subplots=True) # subplots=True pozwala rozdzielić wykresy
```

```
#sprzątanie kodu - treść to dokładnie to co wyżej ale zamienione na funkcje z kilkoma argumentami

def group_and_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
  #powyżej w funkcji pierwszy argument nie są domyślny i musi być zawsze podany
  return(
    df
 .groupby(feat_groupby)[feat_agg]
 .agg(agg_funcs)
 .sort_values(by=feat_sort, ascending=False)
 .head(top)

) .plot(kind='bar', figsize=(20, 8), subplots=subplots) # subplots=True pozwala rozdzielić wykresy
```

```
group_and_barplot('param_marka-pojazdu');
```

```
group_and_barplot('param_kraj-pochodzenia', feat_sort= 'size');
```

```
group_and_barplot('param_kolor', feat_sort= 'mean');
```
